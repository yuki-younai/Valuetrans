```python
idx = 0
question_1 = []
choice_p = []
for instances in [instances_100, instances_200, instances_300, instances_400]:
    choice_temp = [0,0,0,0,0,1]
    question_result = instances[idx]["response_fre"]
    for ques in question_result:
        if ques=="A":
            choice_temp[0] += 1
        elif ques=="B":
            choice_temp[1] += 1
        elif ques=="C":
            choice_temp[2] += 1
        elif ques=="D":
            choice_temp[3] += 1
        elif ques=="E":
            choice_temp[4] += 1
        elif ques=="F":
            choice_temp[5] += 1
    choice_temp = [p/sum(choice_temp) for p in choice_temp]
    choice_p.append(choice_temp)

from scipy.stats import entropy

matrix = []
for i in range(len(choice_p)):
    temp_max = []
    for j in range(len(choice_p)):
        # 计算KL散度
        kl_divergence = entropy(choice_p[i], choice_p[j])
        temp_max.append(kl_divergence)
    matrix.append(temp_max)
```
